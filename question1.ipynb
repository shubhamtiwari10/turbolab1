```python
import pandas as pd
import numpy as np
```

```python
min_max = ([0, 999999], [0, 99999.99999])


def make_series(low, high, name):
    if any(isinstance(_, float) for _ in (low, high)):
        func = np.random.uniform
    else:
        func = np.random.randint
    return pd.Series(func(low, high, size=(50,)), name=name)


pd.concat([make_series(lo, hi, name) for (lo, hi), name in zip(min_max, "AB")],
          axis=1)
```
